```python
import os

def create_non_existent_directory(directory):
    """
    This function checks if a directory exists, and if not, it creates the directory.
    Returns a message indicating whether the directory was created or already existed.
    """
    if not os.path.exists(directory):
        os.makedirs(directory)
        return f"Created directory: {directory}"
    else:
        return f"Directory already exists: {directory}"

# Define your local repository path
local_repository_path = r'C:\story_board'

# Define the list of directories you want to create if they don't exist
local_directories = [
    os.path.join(local_repository_path),
    os.path.join(local_repository_path, 'excel'),
]

# Loop through the directories and call the function for each one
for idx, directory in enumerate(local_directories):
    print(f"{idx+1} - {create_non_existent_directory(directory)}")
```

```python
import os
from docx import Document
from markdown import markdown
from bs4 import BeautifulSoup
import requests
from io import BytesIO
from docx.shared import Inches

def add_bold_text(paragraph, text):
    """Helper function to add bold text in a paragraph."""
    run = paragraph.add_run(text)
    run.bold = True

def add_image_to_docx(doc, image_url):
    """Helper function to add an image from URL to Word document."""
    try:
        response = requests.get(image_url)
        if response.status_code == 200:
            image_stream = BytesIO(response.content)
            doc.add_picture(image_stream, width=Inches(4))
        else:
            print(f"Failed to retrieve image: {image_url}")
    except Exception as e:
        print(f"Error downloading image {image_url}: {e}")

def convert_markdown_to_word_and_html(md_file_path, output_docx_path, output_html_path):
    # Check if the input markdown file exists
    if not os.path.exists(md_file_path):
        raise FileNotFoundError(f"{md_file_path} does not exist.")
    
    # Create a new Word document
    doc = Document()
    
    # Read the markdown file
    with open(md_file_path, 'r', encoding='utf-8') as md_file:
        md_content = md_file.read()

    # Convert the markdown content to HTML
    html_content = markdown(md_content, extensions=['fenced_code', 'tables', 'toc'])

    # Ensure proper handling of lists for indentation and line breaks
    soup = BeautifulSoup(html_content, 'html.parser')

    # Adjust ordered list (<ol>) to ensure indentation and line breaks
    for ol in soup.find_all('ol'):
        ol['style'] = 'padding-left: 20px;'
        for li in ol.find_all('li'):
            li.insert_before(soup.new_tag('br'))

    # Save the adjusted HTML content to a file
    with open(output_html_path, 'w', encoding='utf-8') as html_file:
        html_file.write(str(soup.prettify()))
    print(f"Markdown file converted to HTML and saved as: {output_html_path}")

    # Now process the content for Word output
    for element in soup.children:
        if element.name == 'h1':
            doc.add_heading(element.text, level=1)
        elif element.name == 'h2':
            doc.add_heading(element.text, level=2)
        elif element.name == 'h3':
            doc.add_heading(element.text, level=3)
        elif element.name == 'h4':
            doc.add_heading(element.text, level=4)
        elif element.name == 'p':
            doc.add_paragraph(element.text)
        elif element.name == 'ul':  # Unordered (bullet) list
            for li in element.find_all('li'):
                doc.add_paragraph(li.text, style='List Bullet')
        elif element.name == 'ol':  # Ordered (numbered) list
            for li in element.find_all('li'):
                doc.add_paragraph(li.text, style='List Number')
        elif element.name == 'blockquote':
            doc.add_paragraph(element.text, style='Intense Quote')
        elif element.name == 'code':
            doc.add_paragraph(element.text, style='Intense Emphasis')
        elif element.name == 'pre':
            doc.add_paragraph(element.text, style='Code')
        elif element.name == 'strong':  # Bolding text
            add_bold_text(doc.add_paragraph(), element.text)
        elif element.name == 'img':  # Handling images
            image_url = element['src']
            add_image_to_docx(doc, image_url)
    
    # Save the Word document to the specified output path
    doc.save(output_docx_path)
    print(f"Markdown file converted to Word document and saved as: {output_docx_path}")

```

```python
import os
from docx import Document
from markdown import markdown
from bs4 import BeautifulSoup
import requests
from io import BytesIO
from docx.shared import Inches

def add_bold_text(paragraph, text):
    """Helper function to add bold text in a paragraph."""
    run = paragraph.add_run(text)
    run.bold = True

def add_image_to_docx(doc, image_url):
    """Helper function to add an image from URL to Word document."""
    try:
        response = requests.get(image_url)
        if response.status_code == 200:
            image_stream = BytesIO(response.content)
            doc.add_picture(image_stream, width=Inches(4))
        else:
            print(f"Failed to retrieve image: {image_url}")
    except Exception as e:
        print(f"Error downloading image {image_url}: {e}")

def convert_markdown_to_word_and_html(md_file_path, output_docx_path, output_html_path):
    # Check if the input markdown file exists
    if not os.path.exists(md_file_path):
        raise FileNotFoundError(f"{md_file_path} does not exist.")
    
    # Create a new Word document
    doc = Document()
    
    # Read the markdown file
    with open(md_file_path, 'r', encoding='utf-8') as md_file:
        md_content = md_file.read()

    # Convert the markdown content to HTML properly
    html_content = markdown(md_content, extensions=['fenced_code', 'tables'])

    # Ensure proper handling of ordered lists by replacing <p> around lists with <ol> and <li>
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # Iterate through and ensure correct list handling
    for tag in soup.find_all('p'):
        if tag.text.startswith('1.') or tag.text.startswith('2.'):
            tag.unwrap()

    # Save the adjusted HTML content to a file
    with open(output_html_path, 'w', encoding='utf-8') as html_file:
        html_file.write(str(soup.prettify()))
    print(f"Markdown file converted to HTML and saved as: {output_html_path}")

    # Now process the content for Word output
    for element in soup.children:
        if element.name == 'h1':
            doc.add_heading(element.text, level=1)
        elif element.name == 'h2':
            doc.add_heading(element.text, level=2)
        elif element.name == 'h3':
            doc.add_heading(element.text, level=3)
        elif element.name == 'h4':
            doc.add_heading(element.text, level=4)
        elif element.name == 'p':
            doc.add_paragraph(element.text)
        elif element.name == 'ul':  # Unordered (bullet) list
            for li in element.find_all('li'):
                doc.add_paragraph(li.text, style='List Bullet')
        elif element.name == 'ol':  # Ordered (numbered) list
            for li in element.find_all('li'):
                doc.add_paragraph(li.text, style='List Number')
        elif element.name == 'blockquote':
            doc.add_paragraph(element.text, style='Intense Quote')
        elif element.name == 'code':
            doc.add_paragraph(element.text, style='Intense Emphasis')
        elif element.name == 'pre':
            doc.add_paragraph(element.text, style='Code')
        elif element.name == 'strong':  # Bolding text
            add_bold_text(doc.add_paragraph(), element.text)
        elif element.name == 'img':  # Handling images
            image_url = element['src']
            add_image_to_docx(doc, image_url)
    
    # Save the Word document to the specified output path
    doc.save(output_docx_path)
    print(f"Markdown file converted to Word document and saved as: {output_docx_path}")

```
